```python
# ===================================================
# Raw data : SCHARE_daily_county_ozone_2001_2019
# Purpose Agreegating O3 Data.    
# File Creator : Neha Khairnar
# ==================================================
# 📦 Step 1: Imports
# =========================
import pandas as pd
from pathlib import Path
# =========================
# 📂 Step 2: Load Data
# =========================
BASE_DIR = Path().resolve().parent

#DATA_PATH = BASE_DIR / "data" / "dataset" / "ozone.csv"   # change filename if needed
#df = pd.read_csv("../data/Datasets/SCHARE_daily_county_ozone_2001_2019/SCHARE_daily_county_ozone_2001_2019.csv",  nrows=100)
df = pd.read_csv("../data/Datasets/SCHARE_daily_county_ozone_2001_2019/SCHARE_daily_county_ozone_2001_2019.csv")
#df = pd.read_csv(DATA_PATH)

print("Raw Data Preview:")
display(df.head(10))

# =========================
# 🧹 Step 3: Rename Columns
# =========================
# Adjust if needed
df.columns = [
    "year",
    "date",
    "statefips",
    "countyfips",
    "O3_max_pred",
    "O3_med_pred",
    "O3_mean_pred",
    "O3_pop_pred"
]


# =========================
# 📅 Step 4: Fix Date Format
# =========================
# Example format: 1-Jan-01

df["date"] = pd.to_datetime(df["date"], format="%d%b%Y")

# Extract year safely
df["year"] = df["date"].dt.year

# Convert to standard format YYYY-MM-DD
df["date"] = df["date"].dt.strftime("%Y-%m-%d")

# =========================
# 🧱 Step 5: Format FIPS Codes
# =========================
df["state_fips"] = df["statefips"].astype(str).str.zfill(2)
df["county_fips"] = df["countyfips"].astype(str).str.zfill(3)

# =========================
# 🧾 Step 6: Select Final Columns
# =========================
ozone_clean = df[
    [
        "year",
        "date",
        "state_fips",
        "county_fips",
        "O3_max_pred",
        "O3_med_pred",
        "O3_mean_pred",
        "O3_pop_pred"
    ]
]

# =========================
# 👀 Step 7: Preview Clean Data
# =========================
#print("Cleaned Ozone Data:")
#display(ozone_clean.head(10))


# =========================
# 📊 Step 8: Validate
# =========================
#print("Shape:", ozone_clean.shape)
#print(ozone_clean.info())


# =========================
# 💾 Step 9: Save Output
# =========================
OUTPUT_PATH = BASE_DIR / "data" / "processed" / "neighbourhood_o3_clean.csv"
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
ozone_clean.to_csv(OUTPUT_PATH, index=False)
print("Saved to:", OUTPUT_PATH)

# =========================
# 📊 Step 10: Yearly Aggregation
# =========================
ozone_yearly = (
    ozone_clean.groupby(["year", "state_fips", "county_fips"])
    .agg(
        O3_max=("O3_max_pred", "mean"),
        O3_median=("O3_med_pred", "mean"),
        O3_mean=("O3_mean_pred", "mean"),
        O3_pop=("O3_pop_pred", "mean")
    )
    .reset_index()
)

# =========================
# 👀 Step 11: Preview Yearly Data
# =========================
#print("Yearly Ozone Dataset:")
#display(ozone_yearly.head(10))


# =========================
# 📊 Step 12: Validate
# =========================
#print("Shape:", ozone_yearly.shape)
#print(ozone_yearly.info())


# =========================
# 💾 Step 10: Save Outputs
# =========================

OUTPUT_PATH = BASE_DIR / "data" / "processed" / "neighbourhood_o3_yearly.csv"

OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)

# Save both versions
ozone_yearly.to_csv(OUTPUT_PATH, index=False)

print("Saved to:", OUTPUT_PATH)
```
