```python
# ===================================================
# Raw data : SCHARE_daily_county_pm25_2001_2019
# Purpose Agreegating PM25 Data.    
# File Creator : Neha Khairnar
# ==================================================
# 📦 Step 0: Import Libraries
# ===============================
import pandas as pd
from pathlib import Path
# ===============================
# 📂 Step 1: Load Data
# ===============================
# Adjust path if needed
BASE_DIR = Path().resolve().parent
#DATA_PATH = BASE_DIR /"data"/"dataset"/"SCHARE_daily_county_pm25_2001_2019"/"SCHARE_daily_county_pm25_2001_2019.csv"
#df = pd.read_csv(DATA_PATH, nrows=10)
#df = pd.read_csv("../data/Datasets/SCHARE_daily_county_pm25_2001_2019/SCHARE_daily_county_pm25_2001_2019.csv", nrows=100)
df = pd.read_csv("../data/Datasets/SCHARE_daily_county_pm25_2001_2019/SCHARE_daily_county_pm25_2001_2019.csv")


# Preview raw data
#print("Raw Data:")
#display(df.head(10))

# ===============================
# 🧹 Step 2: Rename Columns (if needed)
# ===============================
# Adjust names based on your actual file
df.columns = [
    "year",
    "date",
    "statefips",
    "countyfips",
    "PM25_max_pred",
    "PM25_med_pred",
    "PM25_mean_pred",
    "PM25_pop_pred"
]

#print("\nColumns after rename:")
#print(df.columns)


# ===============================
# 📅 Step 3: Format Date
# ===============================
# Convert from '1-Jan-01' → datetime standardobject = 2001-01-01
#df["date"] = pd.to_datetime(df["date"], format="%d-%b-%y")
df["date"] = pd.to_datetime(df["date"], format="%d%b%Y")

# Extract year (safe)
df["year"] = df["date"].dt.year


# ===============================
# 🧱 Step 4: Format FIPS Codes
# ===============================
df["state_fips"] = df["statefips"].astype(str).str.zfill(2)
df["county_fips"] = df["countyfips"].astype(str).str.zfill(3)


# ===============================
# 🧾 Step 5: Select Final Columns
# ===============================
pm25_final = df[
    [
        "year",
        "date",
        "state_fips",
        "county_fips",
        "PM25_max_pred",
        "PM25_med_pred",
        "PM25_mean_pred",
        "PM25_pop_pred"
    ]
]


# ===============================
# 👀 Step 6: Preview Clean Data
# ===============================
print("\nCleaned Data:")
display(pm25_final.head(10))



# ===============================
# 📊 Step 7: Basic Validation
# ===============================
#print("\nDataset Shape:", pm25_final.shape)
#print("\nData Types:")
#print(pm25_final.info())


# ======================================================================
# 💾 Step 8: Save Clean Dataset
# =======================================================================
OUTPUT_PATH = BASE_DIR / "data" / "processed" / "neighbourhood_pm25_clean.csv"

# Create folder if not exists
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
pm25_final.to_csv(OUTPUT_PATH, index=False)
print(f"\n✅ File saved to: {OUTPUT_PATH}")

# =====================================================================
# 🧼 Step 9: Create Yearly Aggregation
# =====================================================================
pm25_yearly = (
    pm25_final.groupby(["year", "state_fips", "county_fips"])
      .agg(
          PM25_max=("PM25_max_pred", "mean"),
          PM25_median=("PM25_med_pred", "mean"),
          PM25_mean=("PM25_mean_pred", "mean"),
          PM25_pop=("PM25_pop_pred", "mean")
      )
      .reset_index()
)

# =========================
# 👀 Step 7: Preview Final Output
# =========================
print("Yearly PM2.5 Dataset:")
display(pm25_yearly.head(10))



# =========================
# 📊 Step 8: Validate
# =========================
print("Shape:", pm25_yearly.shape)
print(pm25_yearly.info())


# =========================
# 💾 Step 9: Save Output
# =========================
OUTPUT_PATH = BASE_DIR / "data" / "processed" / "neighbourhood_pm25_yearly.csv"

OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)

pm25_yearly.to_csv(OUTPUT_PATH, index=False)

print("Saved to:", OUTPUT_PATH)
```
