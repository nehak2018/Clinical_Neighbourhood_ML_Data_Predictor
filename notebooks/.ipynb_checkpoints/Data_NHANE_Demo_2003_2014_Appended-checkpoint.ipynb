```python
# ===================================================
# Raw data : 
# Purpose : combining all Demographics data 2004-2014 into one as below
#2003–2004 demographics
#2005–2006 demographics
#...
#2021–2022 demographics
# File Creator : Neha Khairnar
# ==================================================
# 📦 Step 1: Imports
# =========================
import pandas as pd
from pathlib import Path


# =========================
# 📂 Step 2: Define Paths
# =========================
BASE_DIR = Path().resolve().parent
DATA_DIR = BASE_DIR / "data" / "datasets"/ "NHANE_Continuous_DemographicsData_Appended_2003_2014_Data_only"

# List all demographic files (update names as per your files)
FILES = [
    "DEMO_C.csv",
    "DEMO_D.csv",
    "DEMO_E.csv",
    "DEMO_F.csv",
    "DEMO_G.csv",
    "DEMO_H.csv"
]


# =========================
# 📥 Step 3: Load & Append
# =========================
dfs = []

for file in FILES:
    path = DATA_DIR / file
    df = pd.read_csv(path)
    
    # Standardize column names
    df.columns = df.columns.str.upper()
    
    # Optional: Add survey cycle column
    df["CYCLE"] = file.replace(".csv", "")
    
    print(file, "shape:", df.shape)
    
    dfs.append(df)


# =========================
# 🔗 Step 4: Concatenate (Append)
# =========================
demo_all = pd.concat(dfs, axis=0, ignore_index=True)


# =========================
# 👀 Step 5: Preview
# =========================
print("Combined Demographics Preview:")
display(demo_all.head(10))


# =========================
# 📊 Step 6: Validation
# =========================
print("Final shape:", demo_all.shape)

# Check duplicate SEQN (should be unique across cycles)
duplicates = demo_all["SEQN"].duplicated().sum()
print("Duplicate SEQN:", duplicates)


# =========================
# 🧼 Step 7: Optional Cleaning
# =========================
# Drop exact duplicates if any
demo_all = demo_all.drop_duplicates()


# =========================
# 💾 Step 8: Save Output
# =========================
OUTPUT_DIR = BASE_DIR / "data" / "processed"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

demo_all.to_csv(OUTPUT_DIR / "nhanes_demographics_2003_2022_appended.csv", index=False)

print("Saved to:")
print(OUTPUT_DIR / "nhanes_demographics_2003_2022_appended.csv")
```
