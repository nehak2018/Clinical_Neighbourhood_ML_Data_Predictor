```python
# =========================
# 📦 Step 1: Imports
# =========================
import pandas as pd
from pathlib import Path


# =========================
# 📂 Step 2: Load Data
# =========================
BASE_DIR = Path().resolve().parent
DATA_PATH = BASE_DIR / "data" / "dataset" / "ghi.csv"   # change if needed

df = pd.read_csv(DATA_PATH)

print("Raw Data Preview:")
display(df.head(10))


# =========================
# 🧹 Step 3: Rename Columns
# =========================
df.columns = [
    "statefips",
    "countyfips",
    "year",
    "month",
    "day",
    "ghi"
]


# =========================
# 📅 Step 4: Create Date Column
# =========================
df["date"] = pd.to_datetime(
    df[["year", "month", "day"]]
)


# =========================
# 🧱 Step 5: Fix FIPS Codes
# =========================
# State FIPS → 2 digits
df["state_fips"] = df["statefips"].astype(str).str.zfill(2)

# County FIPS currently 4 digits → fix to 3 digits
df["county_fips"] = df["countyfips"].astype(str).str.zfill(3)

# Combine to full 5-digit FIPS (optional but recommended)
df["fips"] = df["state_fips"] + df["county_fips"]


# =========================
# 🧾 Step 6: Clean Daily Dataset
# =========================
ghi_clean = df[
    [
        "year",
        "date",
        "state_fips",
        "county_fips",
        "ghi"
    ]
]

print("Cleaned Daily GHI Data:")
display(ghi_clean.head(10))


# =========================
# 📊 Step 7: Yearly Aggregation
# =========================
ghi_yearly = (
    ghi_clean.groupby(["year", "state_fips", "county_fips"])
    .agg(
        GHI_mean=("ghi", "mean")
    )
    .reset_index()
)


# =========================
# 👀 Step 8: Preview Yearly Data
# =========================
print("Yearly GHI Dataset:")
display(ghi_yearly.head(10))


# =========================
# 📊 Step 9: Validate
# =========================
print("Shape:", ghi_yearly.shape)
print(ghi_yearly.info())


# =========================
# 💾 Step 10: Save Outputs
# =========================
OUTPUT_DIR = BASE_DIR / "data" / "processed"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

ghi_clean.to_csv(OUTPUT_DIR / "ghi_clean.csv", index=False)
ghi_yearly.to_csv(OUTPUT_DIR / "ghi_yearly.csv", index=False)

print("Saved files:")
print(OUTPUT_DIR / "ghi_clean.csv")
print(OUTPUT_DIR / "ghi_yearly.csv")
```
