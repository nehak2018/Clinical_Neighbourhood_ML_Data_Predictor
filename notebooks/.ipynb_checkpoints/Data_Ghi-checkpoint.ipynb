```python
# ===================================================
# Raw data : SCHARE_GHI_Global_Horizontal_Irradiance_1991_2012
# Purpose Agreegating GHI Data.    
# File Creator : Neha Khairnar
# ==================================================
# 📦 Step 1: Imports
# =========================
import pandas as pd
from pathlib import Path

# =========================
# 📂 Step 2: Load Data
# =========================
BASE_DIR = Path().resolve().parent
#DATA_PATH = BASE_DIR / "data" / "dataset" / "ghi.csv"
#df = pd.read_csv(DATA_PATH)

#df = pd.read_csv("../data/Datasets/SCHARE_GHI_Global_Horizontal_Irradiance_1991_2012/Population-Weighted_Global_Horizontal_Irradiance__1991-2012.csv", nrows=100)
df = pd.read_csv("../data/Datasets/SCHARE_GHI_Global_Horizontal_Irradiance_1991_2012/Population-Weighted_Global_Horizontal_Irradiance__1991-2012.csv")

print("Raw Data Preview:")
display(df.head(10))


# =========================
# 🧹 Step 3: Rename Columns
# =========================
df.columns = [
    "statefips",
    "countyfips",
    "year",
    "month",
    "day",
    "ghi"
]


# =========================
# 📅 Step 4: Create Date
# =========================
df["date"] = pd.to_datetime(df[["year", "month", "day"]])

# =========================
# 🧱 Step 5: Fix FIPS (IMPORTANT FIX)
# =========================
# State FIPS → 2 digits
df["state_fips"] = df["statefips"].astype(str).str.zfill(2)

# County FIPS (already combined like 1001)
#df["countyfips"] = df["countyfips"].astype(str).str.zfill(5)

# Extract last 3 digits for county
#df["county_fips"] = df["countyfips"].str[-3:]
df["county_fips"] = df["countyfips"].astype(str).str.zfill(5).str[-3:]

# =========================
# 🧾 Step 6: Clean Dataset
# =========================
ghi_clean = df[
    [
        "year",
        "date",
        "state_fips",
        "county_fips",
        "ghi"
    ]
]

print("Cleaned Daily GHI:")
display(ghi_clean.head(10))

# ======================================================================
# 💾 Step 8: Save Clean Dataset
# =======================================================================

OUTPUT_PATH = BASE_DIR / "data" / "processed" / "neighbourhood_ghi_clean.csv"

# Create folder if not exists
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
ghi_clean.to_csv(OUTPUT_PATH, index=False)
print(f"\n✅ File saved to: {OUTPUT_PATH}")

# =========================
# 📊 Step 9: Yearly Aggregation
# =========================
ghi_yearly = (
    ghi_clean.groupby(["year", "state_fips", "county_fips"])
    .agg(
        GHI_mean=("ghi", "mean")
    )
    .reset_index()
)

# =========================
# 👀 Step 8: Preview
# =========================
print("Yearly GHI:")
display(ghi_yearly.head(10))


# =========================
# 📊 Step 9: Validate
# =========================
print("Shape:", ghi_yearly.shape)
print(ghi_yearly.info())

# =========================
# 💾 Step 10: Save
# =========================
OUTPUT_PATH = BASE_DIR / "data" / "processed" / "neighbourhood_ghi_yearly.csv"

OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)

ghi_yearly.to_csv(OUTPUT_PATH, index=False)

print("Saved to:", OUTPUT_PATH)
```
