```python
# =========================
# 📦 Step 1: Imports
# =========================
import pandas as pd
from pathlib import Path


# =========================
# 📂 Step 2: Paths
# =========================
BASE_DIR = Path().resolve().parent
DATA_DIR = BASE_DIR / "data" / "processed"

pm25 = pd.read_csv(DATA_DIR / "neighbourhood_pm25_yearly.csv")
o3 = pd.read_csv(DATA_DIR / "neighbourhood_o3_yearly.csv")
ghi = pd.read_csv(DATA_DIR / "neighbourhood_ghi_yearly.csv")
omiuv = pd.read_csv(DATA_DIR / "neighbourhood_omiuv_yearly.csv")


# =========================
# 📊 Step 3: Filter Common Years (2004–2012)
# =========================
COMMON_START = 2004
COMMON_END = 2012

pm25 = pm25[(pm25["year"] >= COMMON_START) & (pm25["year"] <= COMMON_END)]
o3   = o3[(o3["year"] >= COMMON_START) & (o3["year"] <= COMMON_END)]
ghi  = ghi[(ghi["year"] >= COMMON_START) & (ghi["year"] <= COMMON_END)]
omiuv= omiuv[(omiuv["year"] >= COMMON_START) & (omiuv["year"] <= COMMON_END)]


# =========================
# 🧹 Step 4: Standardize Column Names
# =========================

# PM2.5
pm25 = pm25.rename(columns={
    "PM25_max": "PM25_max_pred",
    "PM25_median": "PM25_med_pred",
    "PM25_mean": "PM25_mean_pred",
    "PM25_pop": "PM25_pop_pred"
})

# Ozone
o3 = o3.rename(columns={
    "O3_max": "O3_max_pred",
    "O3_median": "O3_med_pred",
    "O3_mean": "O3_mean_pred",
    "O3_pop": "O3_pop_pred"
})

# GHI
ghi = ghi.rename(columns={
    "GHI_mean": "GHI"
})

# OMIUV
omiuv = omiuv.rename(columns={
    "EDD_mean": "OMIUV_edd",
    "EDR_mean": "OMIUV_edr",
    "I305_mean": "OMIUV_i305",
    "I310_mean": "OMIUV_i310",
    "I324_mean": "OMIUV_i324",
    "I380_mean": "OMIUV_i380"
})


# =========================
# 🔗 Step 5: INNER MERGE
# =========================
df = pm25.merge(o3, on=["year", "state_fips", "county_fips"], how="inner")
df = df.merge(ghi, on=["year", "state_fips", "county_fips"], how="inner")
df = df.merge(omiuv, on=["year", "state_fips", "county_fips"], how="inner")


# =========================
# 🧱 Step 6: Rename FIPS
# =========================
df = df.rename(columns={
    "state_fips": "statefips",
    "county_fips": "countyfips"
})


# =========================
# 🧾 Step 7: Select Final Columns
# =========================
final_cols = [
    "year", "statefips", "countyfips",
    
    "PM25_max_pred", "PM25_med_pred", "PM25_mean_pred", "PM25_pop_pred",
    
    "O3_max_pred", "O3_med_pred", "O3_mean_pred", "O3_pop_pred",
    
    "GHI",
    
    "OMIUV_edd", "OMIUV_edr", "OMIUV_i305",
    "OMIUV_i310", "OMIUV_i324", "OMIUV_i380"
]

df_final = df[final_cols]


# =========================
# 👀 Step 8: Preview
# =========================
print("Final Dataset Preview:")
display(df_final.head(10))


# =========================
# 📊 Step 9: Validation
# =========================
print("Shape:", df_final.shape)
print(df_final.info())

# Check missing values (should be minimal or none)
print("\nMissing values:\n", df_final.isnull().sum())


# =========================
# 💾 Step 10: Save Final Dataset
# =========================
OUTPUT_PATH = DATA_DIR / "neighborhood_environmental_2004_2012.csv"

df_final.to_csv(OUTPUT_PATH, index=False)

print("✅ Saved to:", OUTPUT_PATH)
```
